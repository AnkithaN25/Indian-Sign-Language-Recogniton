# Indian Sign Language Detection

End-to-end pipeline using SSD MobileNetV2 transfer learning via the TensorFlow Object Detection API.

**Pipeline:**
1. Create label map
2. Generate TF Records from annotated images
3. Update pipeline config for transfer learning
4. Train the model (run in terminal)
5. Load checkpoint and run real-time webcam inference

> All logic is in `src/`. This notebook is an interactive walkthrough — you can run the full pipeline from the command line without it.
